# 리스트(List)


리스트에는 정보를 저장할 수 있으며 저장할 수 있는 정보량에는 제한이 없습니다.

 리스트는 특정 순서가 있는 항목의 모음입니다. 알파벳 글자, 숫자, 문자열 등 무엇이든 저장가능합니다.
 리스트에는 항목이 보통 두 개 이상 저장되므로 리스트 이름은 letters나 digits, names 처럼 복수형으로 작성합니다. 

파이썬에서 리스트는 대괄호([])로 표현하며, 포함되는 항목들은 콤마로 구분하여 저장합니다. 

```python
animals = ['tiger', 'lion', 'rabbit', 'horse']
print(animals)
```

## 리스트의 항목에 접근

리스트는 순서가 있는 데이터 모음입니다. 따라서 원하는 항목에 접근하기 위해서는 인덱스를 지정해서 접근할 수 있습니다. 항목에 접근하기 위해서는 리스트 이름에 대괄호를 적고 접근하기 원하는 항목의 인덱스를 지정하면 됩니다. 리스트의 인덱스는 0부터 시작합니다.

```python
animals = ['tiger', 'lion', 'rabbit', 'horse']
print(animals[0])
```

리스트에 저장된 항목이 문자열이므로 문자열 클래스가 제공하는 메서드를 각각의 요소에 접근하여 사용할 수 있습니다. 

```python
animals = ['tiger', 'lion', 'rabbit', 'horse']
print(animals[2].title())
```

인덱스 사용시 주의 사항 : 예제의 리스트에는 항목이 4개 이므로 인덱스는 0~3까지 지정됩니다. 이 범위를 넘어서는 인덱스를 사용하면 IndexError 가 발생합니다. 

```python
animals = ['tiger', 'lion', 'rabbit', 'horse']
print(animals[3].title())
```

현재 예제의 리스트의 인덱스는 뒤로부터는 -1 ~ -4까지 사용할 수 있습니다.

```python
animals = ['tiger', 'lion', 'rabbit', 'horse']
print(animals[-1].title())
```

리스트의 각 항목은 일반 변수처럼 사용할 수 있습니다. 예를 들어 -f 문자열을 써서 리스트에 들어 있는 값을 메시지로 만들 수 있습니다. 

```python
animals = ['tiger', 'lion', 'rabbit', 'horse']
message = f"My favorite animal is {animals[0].title()}."

print(message)
```

## 리스트의 항목 변경, 추가, 제거

리스트는 동적 자료구조입니다. 동적이라는 말은 리스트를 만든 뒤에 새 항목을 추가하거나 기존 항목을 제거할 수 있다는 의미입니다. 예를 들어 하늘에 있는 비행선 격추하는 게임을 만든다고 가정해보면 리스트에 비행선을 저장한 다음, 플레이어가 격추하면 리스트에서 해당 비행선을 제거하고, 새 비행선이 나타나면 리스트에 추가합니다. 게임 진행 중에 리스트의 항목 갯수는 늘었다 줄었다 할 것입니다. 

### 리스트의 항목 수정

항목 하나를 수정하는 문법은 리스트 항목에 접근하는 문법과 비슷합니다. 즉 인덱스를 이용할 수 있습니다.

예를 들어, motorcycles 리스트가 있다면 그 항목을 변경해보겠습니다.

```python
motorcycles = ['honda', 'yamaha', 'suzuki']
print(motorcycles)
```

```python
motorcycles[0] = 'ducati'            # 첫 번째 항목을 변경
print(motorcycles)
```

### 리스트에 항목 추가

리스트에 항목을 추가하는 경우는 상당히 많습니다. 예를 들면 어떤 학급에 전학생이 와서 출석부에 이름을 추가하거나 웹사이트에 새 사용자를 등록하는 경우를 들 수 있겠습니다. 파이썬은 여러 가지 방법으로 기존 리스트에 새 데이터를 추가할 수 있습니다.

### append() 메서드를 사용하여 리스트 끝에 항목 추가

```python
motorcycles = ['honda', 'yamaha', 'suzuki']
motorcycles.append('ducati')
print(motorcycles)
```

```python
motorcycles = []         # [] : 빈 리스트

motorcycles.append('honda')
motorcycles.append('yamaha')
motorcycles.append('suzuki')
print(motorcycles)
```

### insert()로 리스트에 항목 추가 

insert() 메서드를 사용하면 새 항목을 리스트의 중간에 추가할 수 있습니다. 새 항목의 인덱스와 값을 지정하면 됩니다.

```python
motorcycles = ['honda', 'yamaha', 'suzuki']

motorcycles.insert(0, 'ducati')
print(motorcycles)
```

```python
motorcycles.insert(2, 'dearim')
print(motorcycles)
```

### 리스트에서 항목 제거

리스트에서 항목을 제거해야 하는 경우도 있습니다. 한 학급에서 학생이 전학을 가거나 웹사이트 사용자가 탈퇴하는 경우에 해당됩니다. 리스트에서 항목을 제거할 때는 값을 기준으로 제거하거나 위치를 기준으로 제거할 수 있습니다. 

#### del 문으로 항목 제거

제거할 항목의 위치를 알고 있다면 del 문을 사용할 수 있습니다.

```python
motorcycles = ['honda', 'yamaha', 'suzuki']

del motorcycles[1]
print(motorcycles)
```

del로 항목을 제거하면 제거한 항목에 접근할 방법이 없습니다.

리스트 클래스의 멤버 pop() 메서드로 항목 제거하기

```python
motorcycles = ['honda', 'yamaha', 'suzuki']

popped_motorcycle = motorcycles.pop()
print(motorcycles)
print(popped_motorcycle)
```

만약 motorcycles 리스트가 각각의 오토바이를 갖게 된 순서라고 가정하면 pop() 메서드를 사용하여 마지막에 갖게 된 오토바이에 관한 문장을 출력할 수 있습니다.

```python
motorcycles = ['honda', 'yamaha', 'suzuki']

last_owned = motorcycles.pop()
print(f"The last motorcycle I owned was a {last_owned.title()}.")
      
```

```python
motorcycles = ['honda', 'yamaha', 'suzuki']

first_owned = motorcycles.pop(0)
print(f"The first motorcycle I owned was a {first_owned.title()}.")
```

### 값으로 항목 제거하기

리스트에서 제거할 항목의 위치는 모르지만 값을 알고 있다면 리스트 클래스의 멤버인 remove() 메서드를 사용할 수 있습니다.

이 방법으로 항목 'ducati'를 리스트에서 제거해 봅니다.

```python
motorcycles = ['honda', 'yamaha', 'suzuki','ducati']

motorcycles.remove('ducati')
print(motorcycles)
```

remove() 메서드를 사용하면 리스트에서 제거된 항목을 저장하여 사용할 수 있습니다.

```python
motorcycles = ['honda', 'yamaha', 'suzuki','ducati']

too_expensive = 'ducati'

motorcycles.remove(too_expensive)
print(motorcycles)
```

```python
print(f"\nA {too_expensive.title()} is too expensive for me.")
```

### 리스트 다루기

#### 전체 리스트에 대해 루프 실행

리스트 항목마다 같은 작업을 실행하며 리스트를 순회해야 할 때가 많습니다. 예를 들어 게임에서 화면에 있는 모든 항목을 같은 양만큼 움직이다던지 통계에서 리스트의 숫자 전체에 대해 같은 통계 작업을 실행하는 일 등이 있습니다. 이 처럼 리스트의 모든 항목에 같은 일을 수행할 때 루프(Loop)를 사용합니다.

다음 리스트에는 마술사들의 이름이 담겨 있습니다. 이 리스트의 각 이름을 출력해봅시다. 매번 리스트에서 이름을 하나씩 뽑아서 출력할 수도 있지만 리스트가 너무 길면 동일한 작업을 너무 많이 해야 하며, 리스트가 바뀔 때마다 코드를 바꿔야 합니다. 이런 문제 점은 반복문인 for문으로 해결 가능합니다.

for 루프를 써서 리스트의 각 이름을 출력해봅시다. 

```python
magicians = ['alice', 'david', 'carolina']

for magician in magicians : # for문의 마지막은 콜른(:)으로 막아야 합니다.
    print(magician)       # 반복 수행할 명령이 for문의 바디 부분으로 들어옵니다.
                          # for문의 바디부분은 반드시 들여쓰기해야 합니다.
                          # for문의 바디부분에 수행할 명령이 여러 개라면 모두 들여쓰기 해야합니다.
```

for문에서 사용되는 임시변수명은 리스트명과 관련된 이름으로 작성하는 것이 좋습니다.

~~~Python
for cat in cats:
for dog in dogs:
for item in list_of_items:
~~~

### 루프 안에서 더 많은 일 하기

앞의 예제를 확장해서 각 마술사에게 훌륭한 마술사였다고 칭찬하는 메시지를 출력해봅시다.

```python
magicians = ['alice', 'david', 'carolina']

for magician in magicians:
    print(f"{magician.title()}, that was a great trick!")
```

리스트에 두 번째 행을 추가해서 각 마술사에게 다음 마술을 기대한다는 메시지를 만들어 봅시다.

```python
magicians = ['alice', 'david', 'carolina']

for magician in magicians:
    print(f"{magician.title()}, that was a great trick!")       # for문이 영향을 주는 영역
    print(f"I can't wait to see your next trick, {magician.title()}.\n")     # for문의 영역
```

```python
magicians = ['alice', 'david', 'carolina']

for magician in magicians:
    print(f"{magician.title()}, that was a great trick!")       # for문이 영향을 주는 영역
    print(f"I can't wait to see your next trick, {magician.title()}.\n")      # for문의 영역
print("Thank you, everyone. That was a great magic show!")
```
