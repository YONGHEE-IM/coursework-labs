## While 문

for문은 반복가능한(iterable) 자료형에서 각 항목을 순회하며 코드 블록을 반복 수행하거나 지정한 범위만큼 코드 블록을 반복 수행합니다. while문은 조건식이 참인 동안 반복을 수행합니다.|

다음 예는 while 문을 이요하여 숫자를 1부터 10까지 출력합니다.

```python
current_number = 1

while current_number <= 10:
    print(current_number, end=' ')
    current_number += 1
```

## 종료값을 사용하여 반복하기 

종료 값을 정의하여 사용자가 종료 값을 입력하기 전까지는 프로그램을 계속 실행하게 만들 수 있습니다.

## 플래그 사용 

만약 프로그램에 여러 조건이 있다면 그중 한가지라도 참 혹은 거짓인 경우 프로그램을 종료해야 한다면 이를 처리하기 위해서는 변수를 정의해서 전체 프로그램의 종료 여부를 판단해야 합니다. 이런 변수를 특별히 플래그라고 합니다. 예를 들어 flag가 True이면 프로그램을 수행하고, False로 변경되는 순간 종료되도록 while 문을 작성할 수 있습니다.

```python
prompt = "\n문장을 입력하세요: "
prompt += "\n종료하려면 'quit'을 입력하세요."

#flag
active = True 
message = " "

while active:
    message = input(prompt)
    if message == 'quit':
        active = False
    else:
        print(message)
```

## break 

조건 테스트에 관계없이 남은 코드를 실행하지 않고 while문을 즉시 빠져나가려면 break 문을 사용합니다. 예를 들어 상용자에게 방문했던 장소를 묻는 프로그램을 작성한다고 가정해봅시다. 사용자가 'quit'를 입력하는 즉시 break 문을 사용하여 while 문을 벗어날 수 있습니다. 

```python
prompt = "\n방문했던 도시 이름을 입렵하세요: "
prompt += "\n(종료하려면 'quit'를 입력하세요.)"

while True:       # while문의 무한 루프
    city = input(prompt)
    if city == 'quit':
        break
    else:
        print(f"{city}에 여행가고 싶어요!")
```

## continue

break 문이 반복문을 완전히 벗어나는 것과 달리 continue문은 만나는 즉시 루프의 시작 지점으로 되돌아갑니다. 예를 들어 1부터 10까지 숫자중 홀수만 출력하는 프로그램을 보겠습니다.

```python
current_number = 0

while current_number < 10:       
    current_number += 1     
    if current_number % 2 == 0: 
        continue
    print(current_number)
```

위 프로그램을 변경해서 1~10사이의 짝수만 출력하세요.

```python
current_number = 1

while current_number < 10:
    current_number += 1
    if current_number % 2 != 0:
        continue
    print(current_number)
        
```

## 리스트와 딕셔너리에서 while문 사용 

### 리스트의 요소를 다른 리스트로 이동 

웹사이트에 새로 가입은 했지만 아직 확인되지 않은 사용자 리스트가 있다고 가정합니다. 이들 사용자를 확인한 다음, 다른 리스트인 확인된 사용자 리스트에 옮기려면 어떻게 해야 할까요? 한 가지 방법은 while 문을 사용해서 확인되지 않은 사용자 리스트에서 사용자를 꺼내서 확인된 사용자 리스트에 추가하는 것입니다.

```python
unconfirmed_users = ['현석', '미영', '금희']     # 확인하지 않은 사용자들
confirmed_users = []     # 확인된 사용자들

while unconfirmed_users:       # 파이썬은 리스트에 데이터가 하나라도 있으면 True로 판단합니다.
    current_users = unconfirmed_users.pop()

    print(f"확인한 사용자: {current_users}")
    confirmed_users.append(current_users)

print("\n확인된 사용자들은 다음과 같습니다.")
print(confirmed_users)
for confirmed_user in confirmed_users:
    print(confirmed_user)


    
```

### 리스트에서 특정 값을 모두 제거

리스트 클래스의 remove() 메서드를 사용하면 리스트에서 특정값을 제거할 수 있습니다. 리스트에 특정 값을 여러 개 존재하는 경우 이를 모두 제거하려면 while문으로 반복하면서 remove() 메서드를 호출하여 제거해야 합니다. 

다음 예에서 문자열 'cat'을 여러 개 포함하는 리스트가 있을 때, while 문을 사용하여 모든 'cat' 문자열을 제거하고 있습니다.

```python
pets = ['dog', 'cat', 'dog', 'goldfish', 'cat', 'rabbit', 'cat']
```

```python
while 'cat' in pets: 
    pets.remove('cat')
```

```python
print(pets)
```

### 사용자가 입력한 값으로 딕셔너리 채우기

while 문을 사용하여 루프가 반복될 때마다 참가자 이름과 응답을 묻는 설문조사 프로그램을 만듭니다. 각 응답은 특정 사용자와 연결해야 하므로 수집한 데이터는 딕셔너리에 저장할 수 있습니다.

```python
responses = {}

# 플래그를 사용하여 프로그램 종류 여부를 결정합니다.
polling_active = True

while polling_active:
    name = input("\n이름을 입력하세요: ")       # 딕셔너리 키로 사용될 데이터
    response = input("\n등산하기 원하는 산 이름은 무엇입니까?")          # 값으로 사용될 데이터

    responses[name] = response    # 딕셔너리에 항목 추가

    repeat = input("다음 사람을 조사하겠습니까? (yes or no)")
    if repeat == 'no':
        polling_active = False

print("\n--------설문 결과---------")

for name, response in responses.items():  
    print(f"{name}님은 {response}를 등산하기 원합니다.")
```

# 함수 

함수<sup>function</sup>란 하나의 특별한 목적의 작업을 수행하기 위해 독립적으로 설계된 **코드의 집합**으로 정의할 수 있습니다. 함수를 사용하는 가장 큰 이유는 반복적인 프로그래밍을 피할 수 있기 때문입니다. 프로그램에서 특정 작업을 여러번 반복해야 할때는 해당 작업을 수행하는 함수를 작성하며 됩니다. 프로그램 개발시 필요할 때마다 미리 작성한 함수를 호출하면 해당 작업을 반복해서 수행할 수 있습니다. 또 프로그램을 여러 개의 함수로 나누어 작성하면, 모듈화로 인해 전체적인 코드의 가독성도 좋아집니다. 프로그램에 문제가 발생하거나 기능의 변경이 필요할 때도 해당 함수만 변경하면 되므로 유지보수성도 좋아집니다. 함수의 크기는 정해진 바는 없지만 하나의 기능을 하나의 함수로 작성하는 것이 좋습니다.

## 함수의 정의 

다음은 사용자에게 환영 인사를 표시하는 greet_user() 함수입니다.

- 함수를 정의할 때는 키워드 def를 사용합니다. 함수명 옆의 소괄호 안에 함수로 들어오는 입력값을 작성할 수 있습니다.(없을수도 있습니다.) 함수정의는 콜론(:)으로 끝납니다.
- def greet_user(): 다음에 들여쓰기 한 행은 함수의 바디가 됩니다. 이는 함수가 수행할 명령들의 집합입니다.
- """큰따옴표 세개"""로 둘러싼 문장은 모통 도큐먼트 스트링 혹 독스트링 이라고 불리는 주석입니다.
- 함수를 실행하기 위해서는 greet_user()와 같이 함수명을 '호출'해야 합니다.
  
~~~python
    def 함수명():
        함수가 수행할 코드1
        함수가 수행할 코드2
        ...
        함수가 수행할 코드n

#호출방법
함수명()
~~~

```python
def greet_user():
    # 주석입니다.
    """
    간단한 환형 인사를 표시합니다.
    주석라인2
    주석라인3
    """
    print("Hello")

greet_user()
```

### 함수에 정보 전달

greet_user() 함수를 수정하여 Hello 다음에 사용자 이름을 추가하여 출력해보겠습니다. def greet_user()의 괄호안에 메개변수(parameter)명을 username으로 작성하여 값을 받는 용도로 사용하겠습니다. 호출시에는 함수에 전달되는 값(인지: argument)를 소괄호안에 적어서 호출합니다. 

```python
def greet_users(username):      # 함수의 정의 
    print(f"Hello, {username}")
```

```python
greet_users("홍길동")

greet_users("김복순")
```

## 매개변수 전달

함수를 정의할 때 매개변수를 여러 개 받도록 정의할 수 있습니다. 파이썬에서 함수에 매개변수를 정의하는 방법은 여러가지입니다. 

함수 정의에서 매개변수를 선언한 순서대로 살인자 값을 넘겨주는 **위치형 매개변수**를 쓸 수 있고, 매개 변수가 변수 이름과 값으로 구성되는 **키워드 매개변수**를 쓸 수도 있고, 리스트나 딕셔너리도 넘길 수 있습니다.

### 위치형 매개변수

함수를 호출하면 파이썬은 넘겨받은 실인자(값)를 함수 정의에 나열된 매개변수와 연결합니다. 가장 단순한 방법은 매개변수가 나열된 순서대로실인지 값을 넘겨주는 것입니다. 이런 식으로 맞춰진 값을 **위치형 매개변수**라고 부릅니다. 

다음 프로그램에서는 반려동물에 관한 정보를 출력하는 함수가 있습니다. 이 함수는 다음과 같이 반려동물의 종류와 이름을 출력합니다.

```python
def describe_pet(animal_type, pet_name):
    """반려동물에 관한 정보를 출력합니다."""
    print(f"\n나는 {animal_type}를 키우고 있습니다.")
    print(f"내가 키우는 {animal_type}의 이름은 {pet_name}입니다.")

# 호출부
describe_pet('강아지', '해피')
```

### 함수 여러 번 호출하기 

함수는 필요한 만큼 여러 번 호출할 수 있습니다. 두 번째 다른 반려동물에 대한 설명을 출력할 때는 describe_pet()을 한 번 더 호출하기만 하면 됩니다.

```python
def describe_pet(animal_type, pet_name):
    """반려동물에 관한 정보를 출력합니다."""
    print(f"\n나는 {animal_type}를 키우고 있습니다.")
    print(f"내가 키우는 {animal_type}의 이름은 {pet_name}입니다.")


describe_pet('강아지', '해피')
describe_pet('고양이', '톰')
```

#### 순서가 중요한 위치형 매개변수

함수를 호출할 때 위치형 매개변수의 순서를 지키지 않으면 예상치 못한 결과가 발생할 수 있습니다. 

### 키워드 매개변수

**키워드 매개변수**는 함수에 넘기는 키-값 쌍입니다. 인자값 안에 이름과 값을 묶으므로 함수에 넘길 때 혼란스러운 상황이 없고, 호출할 때 순서를 걱정할 필요도 없습니다.

### 기본값(default value)

함수를 정의할 때 매개변수의 **기본값**을 정할 수 있습니다. 함수를 호출할 때 실인자값을 넘기면 함수 실행시 넘긴 실인자 값이 사용됩니다. 실인자값을 넘기지 않으면 함수 정의시 설정한 기본값을 사용합니다. 매개변수의 기본값을 정의해두면 함수 호출시 해당 매개변수는 생략가능합니다. 

예를 들어 describe_pet() 호출이 대부분 개에 대한 내용을 출력한다면 animal_type의 기본값을 'dog'로 정할 수 있습니다. 함수를 호출할 때 animal_type에 해당하는 실인자를 넘기지 않아도 기본값인 'dog'로 설정하여 실행가능하므로 편리하게 호출할 수 있습니다. 

기본값은 함수의 뒤쪽 매개변수로 부터 앞쪽으로 이동하며 순서대로 채우는 방식으로 사용해야 합니다. 이 규칙을 지키지 않으면 에러가 발생합니다.

```python
def describe_pet(pet_name, animal_type = '강아지'):
    """반려동물에 관한 정보를 출력합니다."""
    print(f"\n나는 {animal_type}를 키우고 있습니다.")
    print(f"내가 키우는 {animal_type}의 이름은 {pet_name}입니다.")


   
```

```python
describe_pet('해피')
describe_pet('쫑')
describe_pet('멍뭉이')
describe_pet('키티', '고양이')

```
