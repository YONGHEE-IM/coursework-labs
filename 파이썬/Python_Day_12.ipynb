## 랜덤 워크(Random Walk)

랜덤워크는 물리학, 생물학, 화학, 경제학 등의 분야에 쓰입니다. 예를 들어 물 표면에 밀가루 하나가 떨어졌다고 상상해봅시다. 물 분자가 계속 움직이므로, 밀가루 역시 그에 따라 움직일 것입니다. 물의 분자 운동은 랜덤하므로, 그 표면을 따라 흐르는 밀가루의 움직임 역시 랜덤하고 이러한 형태의 움직임을 랜덤워크라고 부를 수 있습니다. 

### RandomWalk 클래스 생성

클래스는 경로의 방향을 랜덤하게 결정합니다. 이 클래스는 세가지 속성(멤버변수)이 필요합니다. 하나는 경로에 사용된 포인트 수를 저장하고, 나머지 둘은 각 포인트의 x와 y좌표를 저장합니다. 

필요한 클래스는 __init__() 메서드와 fill_walk() 메서드 두 가지 입니다. 두 번쨰 메서드에서 경로의 포인트로 사용될 좌표들을 계산합니다. 일단 __init__()메서드를 작성하겠습니다.

```python
from random import choice
import matplotlib.pyplot as plt

class RandomWalk:
    """랜덤 워크를 만드는 클래스"""

    def __init__(self, num_points=5000):
        """경로 속성을 초기화합니다."""
        self.num_points = num_points
        # 모든 경로는(0,0)에서 시작하므로 각각의 좌표를 초기화합니다.
        # 경로는 5000개를 저장해야 하므로 x, y 좌표를 담을 리스트 두 개가 필요합니다.
        self.x_values = [0]
        self.y_values = [0]
        
    # 각각의 좌표를 임의로 뽑아내는 메서드 구현
    def fill_walk(self):
        """경로의 포인트를 모두 계산"""

    # 경로가 원하는 길이가 될 때까지 계속 반복하면 좌표를 구하여 저장합니다.
        while len(self.x_values) < self.num_points:

            # 어느 방향으로 얼마나 갈지 결정
            x_direction = choice([1, -1])    # 리스트 안의 값들중 하나를 뽑아줍니다.
            x_distance = choice([0, 1, 2, 3, 4])    # 이동할 거리를 선택
            x_step = x_direction * x_distance     # 선택한 방향으로 선택한 거리만큼 좌표계산
            
            y_direction = choice([1, -1])
            y_distance = choice([0, 1, 2, 3, 4])
            y_step = y_direction * y_distance

            # 어떤 방향으로도 향하지 않는 경우 다시 선택합니다.
            if x_step == 0 and y_step == 0:
                continue 

            # 새 위치를 계산합니다.    이전 좌표 (3, 5) : x_step = 3, y_step = 2이면
            # 다음에 이동할 위치는 (6, 7)이 됩니다.
            x = self.x_values[-1] + x_step    # 최종적으로 이동한 x좌표로 부터 계산한 방향과 거리만큼의 위치로 이동한 좌표를 계산
            y = self.y_values[-1] + y_step

            self.x_values.append(x)
            self.y_values.append(y)

# 랜덤워크 객체를 생성하고 좌표를 저장(fill_walk()를 호출)한후 그래프로 출력

rw = RandomWalk()
rw.fill_walk()     # 실제 좌표를 선택해서 저장합니다.


# 포인터를 그립니다.
plt.style.use('classic')
fig, ax = plt.subplots()
ax.scatter(rw.x_values, rw.y_values, s=15)
plt.show()
```

### 랜덤 워크 여러개 출력

다음과 같이 이전 코드를 while 루프에 넣어 랜덤 워크를 여러번 만들 수 있습니다.

```python
import matplotlib.pyplot as plt

while True:
    rw = RandomWalk()
    rw.fill_walk()

    # 포인트를 그립니다.
    plt.style.use('classic')
    fig, ax = plt.subplots()
    ax.scatter(rw.x_values, rw.y_values, s=15)
    plt.show()

    keep_running = input("Make another walk? (y/n): ")
    if keep_running == 'n':
        break
```

## 플로틀리로 주사위 굴리기

플로틀리는 브라우저에서 사용할 시각화를 만들 때 특히 유용합니다. 이 시각화는 뷰어의 화면에 자동으로 맞춰지며 플로틀리로 만든 시각화는 대화형이기도 합니다. 즉 사용자가 화면의 특정 요소에 마우스를 올리면 해당 요소에 관한 정보를 출력할 수 있습니다.

프로그램에서는 주사위를 굴린 결과를 분석할 겁니다. 일반적인 주사위는 1부터 6사이의 숫자가 나오고 그 확률은 모두 같습니다. 하지만 주사위를 두 개 굴리면 특정 숫자가 다른 숫자에 비해 자주 나올 가능성이 있습니다. 주사위를 굴리는 행동을 시뮬레이션하고 데이터 세트를 만들고 분석해서 어느 숫자가 가장 많이 나오는지 알아보겠습니다.

### Die 클래스 만들기

다음은 주사위 하나를 굴리는 행동을 나타내는 Die 클래스입니다. 

```python
from random import randint

class Die: 
    """주사위 하나를 나타내는 클래스"""

    def __init__(self, num_sides=6):
        """6면체 주사위를 나타내기 위해 기본값을 설정했습니다."""
        self.num_sides = num_sides
        
    # 주사위를 실제로 굴립니다.
    def roll(self):
        return randint(1, self.num_sides)   # 1 부터 주사위 면수까지의 값증 하나를 임의로 선택해서 반환합니다.

# Die 클래스 타입의 객체를 생성합니다.
die = Die()

results = []

for roll_num in range(100):
    result = die.roll()
    results.append(result)

print(results)
```

### 결과 분석하기

각 숫자가 얼마나 나왔는지 결과를 분석해 보겠습니다. 

```python
die = Die()

results = []

# 주사위를 1000번 굴리고 결과를 리스트에 담습니다.
for _ in range(1000):
    result = die.roll()
    results.append(result)

# 결과를 분석합니다.
frequencies = []
for value in range(1, die.num_sides + 1):
    frequency = results.count(value)
    frequencies.append(frequency)

print(frequencies)
```

값이 여섯가지고 이 중 어떤 값도 다른 값에 비해 눈에 띌만큼 크지 않습니다. 이 결과를 시각화로 만들어 보겠습니다.

### 히스토그램 출력

frequencies 리스트를 사용해 히스토그램을 그릴 수 있습니다. 히스토그램은 각 결과가 얼마나 자주 나타나는지 보여주는 일종의 막대 그래프입니다.

```python
from plotly.graph_objs import Bar, Layout
from plotly import offline

die = Die()

results = []

# 주사위를 1000번 굴리고 결과를 리스트에 담습니다.
for _ in range(1000):
    result = die.roll()
    results.append(result)

# 결과를 분석합니다.
frequencies = []
for value in range(1, die.num_sides + 1):
    frequency = results.count(value)
    frequencies.append(frequency)

# 결과를 시각화 힙니다.
x_values = list(range(1, die.num_sides + 1))
data = [Bar(x=x_values, y=frequencies)]


x_axis_config = {'title' : 'Result'}
y_axis_config = {'title' : 'Frequency of Result'}
# 출력할 그래프의 레이아웃 설정
my_layout = Layout(title='Results of rolling one D6 1000 times', 
                   xaxis = x_axis_config, yaxis = y_axis_config)

offline.plot({'data' : data, 'layout' : my_layout}, filename='d6.html')
```

플로틀리는 range() 함수가 반환하는 값을 직접적으로 받지 않으므로 list() 함수를 써서 명시적으로 리스트로 반환해야합니다. 플로틀리의 Bar() 클래스는 막대 그래프 형태로 표현될 데이터 세트입니다. 이 클래스에는 x값 리스트와 y값 리스트가 필요합니다.

각 축은 여러 가지 방법으로 설정할 수 있으며, 설정 옵션은 딕셔너리에 저장됩니다. Layout() 클래스는 그래프 레이아웃과 설정 전체를 나타내는 객체를 반환합니다. 여기서는 그래프 전체 제목과 x, y 축 설정 딕셔너리를 함께 입력했습니다.

그래프를 만들 때는 offline.pilot() 함수를 호출합니다. 이 함수는 데이터와 레이아웃 객체가 포함된 딕셔너리를 요구하며 그래프를 저장할 파일 이름도 받습니다. 

플로틀리로 만든 시각화는 대화형이라고 했습니다. 그래프의 막대에 마우스를 올리면 관련된 데이터를 보여줍니다. 이 기능은 그래프 하나에 여러 가지 데이터 세트를 표시해야 할때 특히 유용합니다. 

### 주사위 두 개 굴리기

주사윌 두 개를 굴리면 결과 수가 많아지고 분포 또한 변경됩니다. D6 주사위 두 개를 만들어 굴리는 코드로 변경하겠습니다. 

```python
from plotly.graph_objs import Bar, Layout
from plotly import offline

die1 = Die()
die2 = Die()


results = []

# 주사위를 몇 번 굴리고 결과를 리스트에 담습니다. 

for _ in range(1000):
    result = die1.roll() + die2.roll()
    results.append(result)

#결과를 분석합니다.

frequencies = []
max_result = die1.num_sides + die2.num_sides
for value in range(2, max_result + 1):
    frequency= results.count(value)
    frequencies.append(frequency)


# 결과를 시각화합니다.
x_values = list(range(2, max_result + 1))
data = [Bar(x=x_values, y=frequencies)]

x_axis_config = {'title' : 'Result', 'dtick' : 1}
y_axis_config = {'title' : 'Frequency of Result'}

my_layout = Layout(title= 'Result of rolling two D6 dice 1000 times',
                   xaxis= x_axis_config, yaxis=y_axis_config)

offline.plot({'data': data, 'layout' : my_layout}, filename = 'd6_d6.html')
    
    
```
