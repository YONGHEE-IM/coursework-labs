### 의사코드

프로그래밍 언어가 아닌 진짜 언어로 만들 프로그램을 계획하기 위해서는 의사 코드를 사용합니다. 의사코드는 각 행은 하나의 함수 호출처럼 느껴져야 합니다. 그런 다음 나중에 계획이 의미하는 함수를 작성합니다. 

1. 누가 먼저 빈 틱택토 보드를 만들 것인지 결정
2. 아무도 이기지 못하고 동점이 아니며 플레이어가 플레이할 차례라면 말을 위치 시킬 칸을 선택하고 보드를 업데이트
3. 그렇지 않으면 말의 위치를 계산하고 체스판을 업데이트한후 출력하고 다음 턴으로 넘어간다.
4. 승자가 결정되면 축하 메시지를 출력하거나 비겼음을 알립니다.


### 데이터 표현

게임보드를 화면에 인쇄할 것이므로 하나의 문자 "X" 또는 "O"로 말을 표현합니다. 빈 공간은 아직 말을 놓지 않은 공간을 나타냅니다. 각 플레이어가 말을 놓음으로써 보드 데이터가 바뀔 것이므로 보드 자체는 리스트를 사용합니다. 틱택토 보드에는 9개의 정사각형이 있으므로 리스트는 9개의 요소로 구성되어야 합니다. 보드의 각 정사각형은 게임에서의 위치에 해당합니다. 

     0  |  1  |  2
    ----------------
     3  |  4  |  5
    ----------------
     6  |  7  |  8

판 위의 각각의 정사각형이나 위치는 0 - 8 숫자로 표시됩니다. 이는 리스트가 9개의 원소들을 갖게 되고 0 - 8의 위치를 갖는다는 것입니다. 플에이어와 컴퓨터가 하는 플에이어에서는 각각의 위치에 "x"와 "O"로 나타날 것입니다. 현재 턴을 나타내는 변수도 "X"나 "O" 중 하나가 됩니다.

### 게임 설정

전역 상수로 게임에서 사용될 상수값들을 선언합니다.

```python
# 파이썬에서는 따로 상수 개념이 없으므로 대문자 이름으로 지어진 변수는 상수라고 약속합니다.
# 전체 함수에서 접근해서 사용할 수 있는 상수 정의

X = "X"       # 게임에 쓰이는 말이 갖을 수 있는 값을 상수로 정의
O = "o"
EMPTY = " "      # 보드의 빈칸을 나타내는 상수
TIE = "TIE"        # 게임이 비겼을 때 사용할 상수
NUM_SQUARES = 9    # 보드의 칸 수 

```

```python
def display_instructions():
    """게임 설명"""
    print(
    """
    Tic-Tac-Toe.
    게임은 유저와 컴퓨터 간의 대결이 될 것입니다.
    0 - 8의 숫자를 입력하면 말을 위치시킬수 있습니다. 숫자는 그림과 같이 보드 위치에 해당합니다.

    0 | 1 | 2
    ---------
    3 | 4 | 5
    ---------
    6 | 7 | 8

    준비하세요, 이제 곧 게임이 시작됩니다.\n
    """
    )

#사용자로부터 "y"나 "n"를 입력받아서 호출부로 반환합니다.
def ask_yes_no(question):
    """yes 또는 no 중 하나를 선택받아 반환"""
    response = None    # 사용자 입력을 받을 변수
    while response not in ("y", "n"):      #입력받은 문자가 "y" "n"이 아닌경우 다시 입력받음
        response = input(question).lower()   # 입력받은 문자를 소문자로 변경하여 변수에 저장
    return response

#아래 함수는 사용자에게 안내문장을 출력하고 해당 범위 내의 숫자를 사용자에게 입력받아 반환합니다.
def ask_number(question, low, high):
    """범위 안의 숫자를 하나 반환합니다."""
    response = None               
    while response not in range(low, high) :
        response = int(input(question))
    return response

def pieces():
    go_first = ask_yes_no("첫 번째 플레이하시기 원합니까? (y/n): ")

    if go_first == "y":
        print("\n당신이 먼저 플레이 합니다.")
        human = X
        computer = O
    else:
        print("\n컴퓨터가 먼저 플레이 합니다.")
        computer = X
        human = O

    return computer, human

# new_board() :
#     보드(리스트)의 9개 정사각형을 모두 빈 상태로 설정한 새 보드를 만들고 이를 반환합니다.

def new_board():
    """새로운 게임 보드 생성"""
    board = []

    for _ in range(NUM_SQUARES):
        board.append(EMPTY)
    return board

# display_board():
#     함수는 보드에 전달된 내용을 화면에 출력합니다. 보드의 각 요소는 공백문자, "X", "O" 중의 한 문자열이므로 함수안에서 이들을 출력합니다.

def display_board(board):
    """화면에 게임 보드 출력"""
    print("\n\t", board[0], "|", board[1], "|", board[2])
    print("\t", "---------")
    print("\n\t", board[3], "|", board[4], "|", board[5])
    print("\t", "---------")
    print("\n\t", board[6], "|", board[7], "|", board[8])

def legal_move(board):
    """놓을 수 있는 인덱스를 반환합니다."""
    moves = []
    for square in range(NUM_SQUARES):
        if board[square] == EMPTY:
            moves.append(square)
    return moves

# winner(board):
#     보드 정보를 읽어들여 누군가 이긴 쪽이 있는지 체크합니다. 이 때 틱택톡 보드에서 이길 수 있는 경우의 수를 인덱스 세 개의 묶음으로 
#     저장한 후 이를 참조하여 결정합니다.

def winner(board):
    """게임의 승자를 결정합니다"""
    WAYS_TO_WIN = ((0, 1, 2),
                   (3, 4, 5),
                   (6, 7, 8),
                   (0, 3, 6),
                   (1, 4, 7),
                   (2, 5, 8),
                   (0, 4, 8),
                   (2, 4, 6))
    
    for row in WAYS_TO_WIN:
        if board[row[0]] == board[row[1]] == board[row[2]] != EMPTY:
            winner = board[row[0]]
            return winner
    # 보드에 빈자리가 없다면 게임이 끝난 것이고 승부가 갈리지 않았으므로 동점 처리합니다.
    if EMPTY not in board:
        return TIE

    return None

def human_move(board, human):
    """플레이어 턴 처리"""
    # 놓을 수 있는 인덱스를 받습니다.
    legal = legal_moves(board)
    move = None
    #
    while move not in legal:
        move = ask_number("어디에 체크 할 건가요? (0 - 8): ", 0, NUM_SQUARES)
        if move not in legal:
            print("\n그 위치는 이미 체크되어 있습니다. 다른 곳을 선택하세요.\n")
    print("좋습니다....")
    return move

# 컴퓨터 턴 처리
def computer_move(board, computer, human):
    """컴퓨터 턴 처리"""
    # 적절한 자리를 찾기 전까지 시뮬레이션을 해야 하므로 원래 보드 상태를 저장해야 합니다.
    board = board[:]

    # 틱택토의 보드 판에는 최적의 자리가 정해져 있습니다.
    BEST_MOVES = (4, 0, 2, 6, 8, 1, 3, 5, 7)

    print("컴퓨터가 칸을 선택할 것입니다.", end= " ")

    #만약 컴퓨터가 이길 수 있는 자리가 있다면 그 자리를 선택합니다.
    for move in legal_moves(board):
        board[move] = computer
        if winner(board) == computer:
            print(move)
            return move
        # 컴퓨터가 이길 위치를 찾지 못했다면 보드를 원래 상태로 되돌립니다.
        board[move] = EMPTY

    # 상대방이 이길 수 있는 자리가 있다면 그 자리를 막습니다.
    for move in legal_moves(board):
        board[move] = human
        if winner(board) == human:
            print(move)
            return move
        board[move] = EMPTY
    
    # 컴퓨터가 이길 위치도 없고 사람이 이길 위치도 없으므로 최적의 위치가 저장된 BEST_MOVE 튜플에서 놓을 자리를 가져옵니다.
    for move in BEST_MOVES:
        if move in legal_move(board):
            print(move)
            return move

def next_turn(turn):
    """턴을 바꿉니다."""
    if turn == X:
        return O
    else:
        return X

def congrat_winner(the_winner, computer, human):
    """축하 문자열 출력"""
    if the_winner != TIE:
        print(f"{the_winner}가 이겼습니다!\n")
    else:
        print("무승부입니다!\n")
    if the_winner == computer:
        print("내 예측대로 당신에게 한번 더 승리했습니다. \n" \
              "컴퓨터가 인간보다 우수하다는 것을 증명했습니다.")
    elif the_winner == human:
        print("내가 인간에게 질 줄이야... 이럴 리 없어...\n" \
              "맹세코 다시는 지지 않아!")
    elif the_winner == TIE:
        print("나와 비기다니 당신은 운이 좋았습니다. \n" \
              "안타깝지만 그것이 당신의 최고 성적일 것입니다.")



def main():
    # 게임정보 출력
    display_instructions()
    # 말을 선택
    computer, human = pieces()
    turn = X
    # 게임 보드 생성
    board = new_board()
    # 게임 보드 출력
    display_board(board)

    # 게임루프
    while not winner(board):
        # 사람이 놓을 턴이면 그 처리를 해줍니다.
        if turn == human:
            move = human_move(board, human)
            board[move] = human
        # 컴퓨터가 놓을 턴이면 그 처리를 해줍니다.
        else:
            move = computer_move(board, computer, human)
            board[move] = computer
        # 이번 턴의 말을 놓았고 보드 정보가 갱신됐으므로 화면도 갱신합니다.
        display_board(board)
        turn = next_turn(turn)

    the_winner = winner(board)
    congrat_winner(the_winner, computer, human)

main()
input("\n\n종료하려면 아무 키나 누르세요.")
            
    
```

# 클래스(Class)

**객체 지향 프로그래밍**(Object Oriented Programming:OOP)은 실세계를 해석하여 소프트웨어를 개발하고자 할 때 관점의 주된 대상을 실세계의 객체 위주로 이해하는 것입니다. 과거의 구조적 프로그래밍 방식(절차지향 프로그래밍)은 실세계의 시스템을 이해할 때 작업의 흐름으로 이해했습니다.

객체지향은 소프트웨어 관점에서 보면 자료구조(멤버 함수)와 그에 대한 연산(메서드)를 묶어서 구성되는 객체들을 정의하고 이들을 적절히 조작함으로써 소프트웨어 구조를 구성하는 것을 말합니다. 따라서 객체 지향 기법은 구조적 기법에 비해 자료와 그를 처리하는 연산의 연결 강도가 높습니다. 

객체지향 프로그래밍에서는 클래스를 만들고 이들 클래스를 바탕으로 객체를 생성합니다. 클래스를 만들 때 객체의 전체 카테고리가 가질 수 있는 일반적 동작을 정의합니다. 

클래스를 바탕으로 객체를 생성하는 것을 인스턴스화라고 부르며 이때 생성되는 객체를 인스턴스라고 부르기도 합니다.

클래스는 복잡한 문제를 단순하게 해주고 공동 작업에 편리하게 쓰입니다.

## 클래스를 만들고 사용하기

Dog 클래스를 작성하겠습니다. Dog 클래스는 실제 존재하는 특정 개가 아니라 단순히 개라는 대상을 표현하기 위한 틀입니다. 클래스를 바탕으로 객체를 생성했을 때(클래스 타입의 변수선언) 생성한 개에 대한 데이터가 메모리에 적재되어 값을 저장하거나 불러올 수 있는 상태가 됩니다. 

우리가 만들 강아지 클래스는  이름과 나이라는 정보를 저장하고 행동으로는 앉기 구르기 등을 할 수 있습니다. 이 때 클래스에 저장되는 이름, 나이등을 멤버변수라고 부르고 앉기 구르기등의 기능을 멤버함수 혹은 메서드라고 부릅니다.

```python
class Dog:
    """개를 모델화"""
    # 클래스는 생성자(__init__)라는 특수한 멤버함수를 갖습니다.
    # 생성자는 객체가 생성될 때 자동으로 수행이 됩니다. 생성자 안에서는 객체의 멤버변수를 초기화 하는 일을 합니다.
    # 생성자의 첫 번째 매개변수는 self이며 이는 메모리에 생성되는 객체를 참조하는 변수입니다.
    def __init__(self, name, age): 
        self.name = name     # 멤버 변수 self.name 선언
        self.age = age       # 멤버 변수 self.age 선언

    def sit(self):           # 멤버 함수(메서드) 
        print(f"{self.name} is now sitting.")

    def roll_over(self):     # 멤버 함수(메서드)
        print(f"{self.name} rolled over!")

my_dog = Dog('Willie', 5)
your_dog = Dog('Lucy', 3)
```

```python
my_dog = Dog('해피', 5) 
print(f"My dog's name is {my_dog.name}.")
print(f"My dog is {my_dog.age} years old.")
```

```python
my_dog.sit()
my_dog.roll_over()
```

### 클래스를 통해서 다수의 인스턴스 생성
클래스의 인스턴스는 원하는 만큼 만들 수 있습니다. 두 마리의 개를 생성해 보겠습니다.

```python
class Dog:
    """개를 모델화"""
    # 클래스는 생성자(__init__)라는 특수한 멤버함수를 갖습니다.
    # 생성자는 객체가 생성될 때 자동으로 수행이 됩니다. 생성자 안에서는 객체의 멤버변수를 초기화 하는 일을 합니다.
    # 생성자의 첫 번째 매개변수는 self이며 이는 메모리에 생성되는 객체를 참조하는 변수입니다.
    def __init__(self, name, age): 
        self.name = name     # 멤버 변수 self.name 선언
        self.age = age       # 멤버 변수 self.age 선언

    def sit(self):           # 멤버 함수(메서드) 
        print(f"{self.name} is now sitting.")

    def roll_over(self):     # 멤버 함수(메서드)
        print(f"{self.name} rolled over!")

my_dog = Dog('Willie', 5)
your_dog = Dog('Lucy', 3)

print(f"My dog's name is {my_dog.name}.")
print(f"My dog is {my_dog.age} years old.")
my_dog.sit()
your_dog.roll_over()
print(f"Your dog's name is {your_dog.name}.")
print(f"Your dog is {your_dog.age} years old.")
```
