# 변수와 기본 타입

변수를 이용하여 문자열을 출력해 봅니다.

프로그램은 message라는 변수를 추가했습니다. 변수는 대입문(=)의 오른쪽에 있는 값과 연결되어 있습니다. 인터프리터(해석기)는 message라는 변수에 "Hello world"라는 텍스트를 연결합니다. 
print()는 문자열을 화면에 출력하는 기능을 갖는 내장함수(파이썬 안에 포함되어 있음)입니다.

```python
# 주석 : 프로그램에 대한 설명을 작성하기 위한 일종의 메모입니다.
# 파이썬 주석은 # 기호를 입력하여 작성합니다.
# 주석은 프로그램 실행시 기계어로 변환되지 않습니다.

# '=' 등호 기호는 프로그래밍에서는 오른쪽의 값을 왼쪽의 변수에 저장을 하거나 연결 시킵니다.
# = 기호를 대입연산자(assignment)라고 부릅니다.

message = "Hello world!"   # message는 변수입니다.
print(message)

# 우리가 message를 변수라고 부르는 이유는 변수가 가지는 값을 변경 가능하기 때문입니다.
message = "Hello everyone!"
print(message)
```

### 문자열 안에서 변수 사용

때로는 문자열 안에 변수 값을 써야 할 때도 있습니다. 예를 들어 변수 하나는 이름, 다른 변수는 성을 각각 지정하여 이들을 합쳐서 전체 이름으로 사용할 수 있습니다.

```python
first_name = "taylor"
last_name = "swift"

full_name = f"{first_name} {last_name}"
print(full_name)
```

이런 문자열을 f-문자열이라고 부릅니다. 여기서 f는 format, 즉 형식을 갖춘다는 의미입니다.


f-문자열을 써서 메시지를 조합하고, 전체 메시지를 다시 변수에 할당할 수도 있습니다.

```python
first_name = "taylor"
last_name = "swift"
full_name = f"{first_name} {last_name}"

message = f"Hello, {full_name.title()}!"
print(message)
```

f-문자열은 파이썬 3.6에서 도입되었으므로 그 이상 버전에서만 사용할 수 있습니다.

이전에 사용하던 방식

```python
first_name = "taylor"
last_name = "swift"
full_name = "{} {}".format(first_name, last_name)

print(full_name)
```

### 탭이나 줄바꿈 문자를 써서 문자열에 공백 추가(이스케이프 시퀀스(탈출문자)라고 부릅니다.)

프로그램에서 공백은 스페이스, 탭, 줄바꿈 기호처럼 화면에 표시되지 않는 문자를 가리킵니다. 공백은 보통 출력 결과를 일정한 형식으로 정리해서 사용자가 읽기 쉽게 만들 때 사용합니다.

```python
print("Python")
```

```python
print("\tPython")
```

문자열에 줄바꿈을 추가하려면 다음과 같이 줄바꿈 문자(개행문자: newline) \n를 사용합니다.

```python
print("Languages:\nPython\nC\nJavaScript")
```

하나의 문자열에 탭과 줄바꿈 문자를 조합해서 사용할 수 있습니다. 문자열 "\n\t"는 먼저 줄을 바꾼후 탭을 넣으라는 의미입니다.

```python
print("Languages:\n\tPython\n\tC\n\tJavaScript")
```

### 문자열 안의 공백 제거

두 문자열을 비교해서 둘이 같은 문자열인지 확인해야 할 때가 많습니다. 예를 들어 웹사이트에 로그인하면서 이름을 입력하는 경우가 있습니다. 사용자가 입력하는 데이터는 공백이 포함되어 있는 경우가 있으며 이 경우 공백을 제거하는 쉬운 방법이 있습니다.

파이썬은 문자열의 오른쪽과 왼쪽 공백을 찾을 수 있습니다. 문자열의 오른쪽에 있는 공백을 모두 제거할 때는 문자열 클래스의 rstrip()메서드를 사용합니다.(rstrip() : right strip())

```python
favorite_language = 'python '
favorite_language
```

```python
favorite_language.rstrip()   # 원본은 변경하지 않습니다.
```

```python
favorite_language
```

```python
# 만약에 공백이 제거된 문자열이 프로그램 내부에서 필요한 경우 변수에 저장을 해야 합니다.
stripped_string = favorite_language.rstrip()
```

```python
stripped_string
```

```python
favorite_language
```

```python
favorite_language = favorite_language.rstrip()
```

```python
favorite_language
```

### 문자열의 왼쪽 공백들을 제거하려면 문자열 클래스의 lstrip() 메서드를 사용합니다.

```python
favorite_language = '          Python'
favorite_language.lstrip()
```

```python
favorite_language
```

### 왼쪽과 오른쪽에 모두 공백이 있는 경우 이를 제거하기 위해서는 strip()를 사용합니다.
