### 여러개의 딕셔너리들이 리스트에 중첩된 경우

게임상에서 현재 화면에 외계인이 여러 명 생성되어야 한다면 생각해 볼 수 있는 방법중 하나는 외계인들에 대한 정보는 딕셔너리에 저장을 하고 이들을 리스트에 담아서 관리하는 것입니다. 다음 예는 각각의 외계인 정보를 딕셔너리에 저장한후 리스트 aliens에 저장하는 경우입니다. 

```python
alien_0 = {'color' : 'green', 'points' : 5}
alien_1 = {'color' : 'yellow', 'points' : 10}
alien_2 = {'color' : 'red', 'points' : 15}

aliens = [alien_0, alien_1, alien_2]

for alien in aliens:
    print(alien)
```

### 외계인 30인을 생성하고 그 중 앞으로부터 5명의 정보를 출력하는 예

```python
# 생성될 외계인을 담을 리스트를 빈리스트로 생성합니다.
aliens = []

for _ in range(30):       # _ : for문안에서 순회중인 숫자를 사용하지 않기 위해서 임시변수를 생성하지 않고 생략하기 위한 기법
    new_alien = {'color' : 'green', 'points' : 5, 'speed' : 'slow'}
    aliens.append(new_alien)

print(aliens)

for alien in aliens[:5]:
    print(alien)
    print("...")
```

게임 진행중 외계인의 일부가 가지는 속성을 변경되어야 하는 경우에는 다음과 같이 for문과 if문을 이용해서 속성을 변경할 수 있습니다. 예제에서는 앞에서부터 세명의 속성을 변경합니다. 

```python
aliens = []

for _ in range(30):       # _ : for문안에서 순회중인 숫자를 사용하지 않기 위해서 임시변수를 생성하지 않고 생략하기 위한 기법
    new_alien = {'color' : 'green', 'points' : 5, 'speed' : 'slow'}
    aliens.append(new_alien)

for alien in aliens[:3]:
    if alien['color'] == 'green':
        alien['color'] = 'yellow'
        alien['points'] = 10
        alien['speed'] = 'medium'
        
for alien in aliens[:5]:
    print(alien)
```

### 딕셔너리 안에 리스트 포함

딕셔너리 안에 리스트를 포함시킬 수 있습니다. 다음 예에서는 피자 주문시 토핑의 종류가 여러가지가 들어긱도 하므로 리스트에 담아서 딕셔너리의 항목으로 추가하고 있습니다. 리스트에 담긴 토핑 정보가 필요한 경우 딕셔너리명 다음의 대괄호 안에 키를 적어 접근할 수 있습니다.

```python
pizza = {
    'crust' : 'thick',
    'toppings' : ['mushrooms', 'extra cheese']
}

print(f"당신은 {pizza['crust']} 두께의 피자를 다음과 같은 토핑을 추가하여 주문했습니다.")
print("주문한 토핑: ")

for topping in pizza['toppings']:
    print("\t" + topping)
```

```python
favorite_languages = {
    '미영' : ['python', 'java'],
    '철민' : ['c'],
    '경호' : ['java', 'c'],
    '희영' : ['python', 'c++']
}

for name, languages in favorite_languages.items():
    print(f"\n{name}가 선호하는 프로그램이 언어: ")
    for language in languages:
        print(f"\t{language.title()}")
```

# 사용자 입력과 while 루프

대부분의 프로그램은 최종 사용자의 문제를 해결하기 위해 만들어집니다. 사용자의 문제를 해결하기 위해서는 사용자로부터 정보를 넘겨 받아야 합니다. 예를 들어 자신이 설문에 응할 수 있는 나이인지 알고 싶다면 프로그램은 사용자의 나이를 알아야 합니다. 따라서 프로그램은 사용자의 나이를 입력 받아야 합니다.

파이썬에서는 **input()** 함수로 사용자로부터 정보를 입력받습니다. 

## input() 함수 동작법

input() 함수는 사용자 입력이 들어올 때까지 대기하다가 사용자 입력을 받으면 입력받은 정보를 변수에 할당하는 식으로 사용합니다. 다음 예제는 사용자로부터 텍스트 입력을 받고 변수에 저장한뒤 출력하고 있습니다.

input 함수는 매개변수가 하나만 있습니다. 이 매개변수는 사용자가 해야할 행동을 설명하기 위한 프롬프트가 됩니다.


```python
message = input("문장을 입력하세요 : ") 
print(message)
```

때때로 여러 줄의 프롬포트가 필요할 수 있습니다. 이런 경우 프롬포트를 저장할 변수를 선언하고 해당 변수에 문자열을 연결시켜서 저장한 다음 input 함수의 입력으로 사용하면 됩니다. 

```python
prompt = "당신이 누구인지 입력하시면 인사 문구를 보내드립니다."
prompt += "\n이름을 입력하세요:"

name = input(prompt)   # input 함수는 입력받은 데이터를 무조건 문자열 형태로 반환합니다.

print(type(name))

print(f"\n안녕하세요, {name}님 즐거운 하루 보내세요.")
```

## int() 함수를 사용하여 숫자로 입력 받기
파이썬은 input() 함수로 입력 받은 데이터를 모두 문자열로 간주합니다.

```python
age = input("나이를 입력하시오:")

age = int(age)

age += 5

print(age)
print(type(age))
```

```python
age = input("나이를 입력 : ")

age = int(age)

if age > 18 :
    print("당신은 투표 가능한 나이입니다.")
else:
    print("당신은 투표 가능한 나이가 아닙니다.")
```

### 나머지 연산자 (Modulo Operator)
나머지 연산자는 첫 번째 피연산자를 두 번째 피연산자로 나눈후의 나머지를 반환합니다.

```python
print(10 % 3)
```

```python
print(7 % 4)
```

나머지 연산자를 사용하여 어떤 숫자가 짝수인지 홀수인지 판단할 수 있습니다.


```python
number = int(input("숫자를 입력하세요 : "))

if number % 2 == 0:
    print(f"\n입력하신 {number}는 짝수입니다.")
else:
    print(f"\n입력하신 {number}는 홀수입니다.")
```

```python
number = int(input("숫자를 입력하세요 : "))

if number % 3 == 0:
    print(f"\n입력하신 {number}는 3의 배수입니다.")
else:
    print(f"\n입력하신 {number}는 3의 배수가 아닙니다.")
```
