### 숫자형 리스트 만들기

숫자 리스트는 예를 들어 게임에서 캐릭터의 위치를 추적해야 한다거나, 플레이어의 최고 기록을 저장해야 하는 경우에 사용할 수 있습니다.

리스트는 숫자를 저장하기 이상적이며 숫자 리스트가 사용할 수 있는 함수가 많이 있습니다. 



#### range() 함수

파이썬의 range() 함수는 연속된 숫자를 쉽게 생성할 수 있습니다.

```python
for value in range(1, 5):
    print(value)
```

주의 해야 하는 것은 range(시작_숫자, 끝_숫자)에서 끝_숫자는 포함되지 않습니다.

숫자 1부터 5까지 생성해서 출력하려면 다음과 같이 작성합니다.

```python
for value in range(1, 6):
    print(value)
```

range 함수는 시작+숫자를 생략할 수 있습니다. 생략한 경우에는 시작되는 숫자가 0으로 자동 설정됩니다.

```python
for value in range(6):
    print(value)
```

#### range()로 숫자 리스트 만들기

숫자 리스트를 원한다면 list() 함수를 써서 range() 결과를 바로 리스트로 만들 수 있습니다. range() 함수를 호출하면서 list() 함수로 감싸면 결과가 숫자 리스트가 됩니다.

```python
numbers = list(range(1, 6))
print(numbers)
for number in numbers:
    print(number)
```

range() 함수를 써서 주어진 범위의 숫자를 건너뛸 수 있습니다. range() 에 세 번째 매개 변수를 넘기면 이 세 번째 매개변수만큼을 더해 숫자를 생성합니다. 

예를 들어 다음은 1부터 10사이의 짝수만 리스트로 만드는 프로그램입니다. 

```python
even_numbers = list(range(2, 11, 2))
print(even_numbers)
```

파이썬에서는 애스터리스크 두 개(**)가 지수를 나타냅니다. 1부터 10까지의 제곱수 리스트를 만들기 위해서는 다음과 같이 작성합니다. 1부터 10까지의 제곱수를 리스트에 담습니다.

```python
squares = []

for value in range(1, 11):
    square = value ** 2                #  square         squares 
                                       #    1              [1]   
                                       #    4              [1, 4]
                                       #    9              [1, 4, 9]
    squares.append(square)

print(squares)
```

```python
squares = []

for value in range(1, 11):
                                       #  square         squares 
                                       #    1              [1]   
                                       #    4              [1, 4]
                                       #    9              [1, 4, 9]
    squares.append(value ** 2)

print(squares)
```

### 리스트 내포(List comprehension)

리스트 내포를 사용하면 한 줄의 코드로 위와 같은 리스트를 생성할 수 있습니다. 리스트 내포는 for루프와 새 항목 생성을 한 행에 결합하여 각 항목을 자동으로 리스트에 추가합니다.

```python
squares = [value ** 2 for value in range(1, 11)]
print(squares)
```

슬라이스를 만들 때는 인덱스가 두 개 필요합니다. 첫 번째 인덱스는 시작하는 항목의 인덱스입니다. 주의할 점은 두 번째 인덱스는 끝나는 항목의 인덱스가 아닌 바로 앞의 인덱스가 됩니다.

```python
players = ['charies', 'matina', 'michael', 'florence', 'eli']
# 인덱스       0         1          2           3        4


print(players[0:3])

# 슬라이스한 데이터가 필요하면 변수에 담으면 됩니다.
sliced_players = players[0:3]
print(players)
print(sliced_players)
```

```python
print(players[1:4])
```

```python
players[:4]
```

# 튜플 (Tuple)

리스트는 프로그램 실행 주에 바뀔 수 있는 항목을 저장하는데 알맞습니다. 리스트(mutable)를 수정할 수 있다는 사실은 웹사이트의 사용자 리스트나 사용자 리스트나 게임의 캐릭터 리스트를 다룰 때 중요합니다. 가끔은 항목을 바꿀 수 없는 리스트가 필요할 때도 있습니다. 이 때 튜플을 사용합니다. 파이썬의 바뀔수 없는 값을 **불변적** (immutable) 이라 부릅니다.

### 튜플 정의

튜플은 리스트의 대괄호([]) 대신에 소괄호 (())를 사용합니다. 튜플을 정의하면 리스트와 마찬가지로 각 항목에 대해 인덱스로 접근 가능합니다. 

예를 들어 크기가 항상 일정한 규격의 사각형이 있다면, 그 사격형의 가로, 세로 크기를 튜플에 저장할 수 있습니다. 

```python
dimensions = (200, 50)
print(f"가로 길이 : {dimensions[0]}")
print(f"세로 길이 : {dimensions[1]}")
```

```python
# 튜플의 항목 값을 변경하려고 시도하면 에러가 발생합니다.
dimensions[0] = 300
```

튜플의 항목을 변경하는 것은 불가능하지만 튜플을 가리키는 변수의 값(대상 객체)은 변경 가능합니다.

```python
dimensions = (300, 70)
print(dimensions)
```

튜플 정의시 주의해야 할 것은 튜플은 항목이 단 하나만 들어 있을 경우 다음과 같이 끝에 콤마를 붙여야 합니다.

```python
my_tuple = (3,)
print(my_tuple)
```

```python
other_dimensions = 150, 20     # 이와 같이 콤마로 구분한 값을 변수에 대입하면 변수 타입이 튜플이 됩니다.
```

```python
type(other_dimensions)
```

# 제어문:

## if 문 (분기문)
어떤 조건에 따라서 여러가지 선택사항중 하나를 고를 때 사용합니다.

```python
cars = ['audi', 'bmw', 'subaru', 'toyota']

for car in cars : 
    if car == 'bmw':            # 현재 순회중인 차 브랜드가 'bmw'인 경우만 대문자로 변경합니다.
                                # 분기문에 사용되는 조건식은 True 혹은 False중 하나의 값만 갖습니다.
                                # 분기문은 조건식이 True인 경우에만 들여쓰기한 블럭을 수행합니다.
                                # if 조건식이 참이라서 블럭이 수행되었다면 아래 else 구문은 수행하지 않습니다.
        print(car.upper())
    else:                       # car != 'bmw'라는 의미 ('bmw'가 아닌경우)
        print(car.title())
```

# 조건 테스트

### 여러 조건 체크

동시에 여러 조건을 체크할 수 있습니다. 예를 들어 조건이 두 가지 있고 두 조건 모두 True일때만 어떤 행동을 취해야 하는 경우가 있습니다. 경우에 따라 조건 중 하나만 True여도 행동을 취해햐 하는 경우도 있습니다. 키워드 and와 or는 이런 상황에 사용하기 적합합니다.

### 모든 조건을 만족해야만 하는 and

두 조건이 모두 True인지 체크하려면 두 조건을 and 키워드로 연결합니다. 예를 들어 두 사람이 모두 21세 이상인지 체크하려면 다음과 같이 작성할 수 있습니다.

```python
age_of_person1 = 32
age_of_person2 = 15

age_of_person1 >= 21 and age_of_person2 >= 21
#    True            and        False             == False
```

```python
age_of_person2 = 25

age_of_person1 >= 21 and age_of_person2 >= 21
```

### 조건 하나만 만족해도 True를 반환하는 or 연산자

키워드 or는 여러 조건을 체크할 때 사용되며, 조건 중 하나만 True여도 전체 테스트 결과가 True가 됩니다. or는 모든 조건이 False인 경우에만 최종 테스트 결과가 False가 됩니다.

```python
age_of_person1 = 32
age_of_person2 = 15

age_of_person1 >= 21 or age_of_person2 >= 21
     True                   False          == True
```

### 값이 목록에 있는지 확인 (in)

때로는 목록에 특정 값이 포함되어 있는지 확인해야 할 수도 있습니다. 예를들어, 웹 사이트에서 누군가의 등록을 완료하기 전에 현재 사용자 이름 목록에 새 사용자 이름이 이미 있는지 확인할 수 있습니다.

특정 값이 목록에 있는지 확인하려면 in 키워드를 사용합니다.

예를 들어 피자 가게에서 고객이 피자에 대해 요청한 토핑 목록을 만든 다음 특정 토핑이 목록에 있는지 확인할 수 있습니다.

```python
requested_toppings = ['mushrooms', 'onions', 'pineapple']
'mushrooms' in requested_toppings
```

```python
if 'pepperoni' in requested_toppings:
    print("페페로니가 있습니다.")
else:
    print("토핑리스트에 페페로니가 없습니다.")
```

### 값이 목록에 없는지 확인

이 상황에서는 not in 키워드를 사용하여 평가합니다.

예를 들어, 포럼에서 논평하는 것이 금지된 사용자 목록이 있다면, 특정 사용자가 논평을 제출하도록 허용하기 전에 사용자가 금지 상태인지 확인할 수 있습니다.

```python
banned_users = ['andrew', 'carolina', 'advid']
user = 'marie'

if user not in banned_users:
    print(f"{user.title()}, you can post a response if you wish.")
```
